```python
def shift_forw(l):
    new_l = l.copy()
    new_l.insert(0, new_l.pop())
    return new_l
def shift_back(l):
    new_l = l.copy()
    new_l.append(new_l.pop(0))
    return new_l
```

```python
polyline_good = (0x1, 0x2, 0x4, 0x6, 0x7, 0x8, 0x9, 0xa, 0xb, 0xd, 0xe, 0x16, 0x1a, \
                 0x1b, 0x1c, 0x27, 0x28, 0x29, 0x2a, 0x2b, 0x32, 0x33, 0x34, 0x35)
polyline_bad = (0x3, 0x15, 0x18, 0x19, 0x1e, 0x1f)
```

```python
def rel_dist(x, y):
    return math.sqrt(x*x + y*y)
```

```python
def check_qout(x, y, quot_cur):
    if (quot_cur == 1):
        return (x >= 0) and (y >= 0)
    if (quot_cur == 2):
        return (x <= 0) and (y >= 0)
    if (quot_cur == 3):
        return (x <= 0) and (y <= 0)
    if (quot_cur == 4):
        return (x >= 0) and (y <= 0)
```

```python
def send_otr(a1, a2, b1, b2, type_line):
    a1 -= cur[0]
    a2 -= cur[1]
    b1 -= cur[0]
    b2 -= cur[1]
    if (rel_dist(a1, a2) > dist_to_wpt) and (rel_dist(b1, b2) > dist_to_wpt):
            return
    for i in range(num_angle):
        k_cur = math.tan((i * 360 / num_angle + small_decline)*math.pi/180)
        quot_cur = math.trunc(i * 360 / num_angle / 90) + 1
        delta_angle = math.tan(math.pi/180)
        if a1 == b1:
            k_otr = 1000
        else: k_otr = (a2 - b2) / (a1 - b1)
        #if abs(k_otr - k_cur) < delta_angle:
            #nam tuda + proverka na chetvert
        b_otr = a2 - k_otr * a1
        x = b_otr / (k_cur - k_otr)
        y = k_cur * x
        if (not check_qout(x, y, quot_cur)):
            continue
        dist_from_cur = rel_dist(x, y)
        if (dist_from_cur > dist_to_wpt):
            continue
        if (type_line in polyline_good) and (cur_type <= 0):
            if math.isnan(df.dist_good[i]) or (df.dist_good[i] > dist_from_cur):
                df.dist_good[i] = dist_from_cur
                if (not math.isnan(df.dist_bad[i])) and (df.dist_bad[i] > df.dist_good[i]):
                    df.dist_bad[i] = np.nan
        if (type_line in polyline_bad) and (cur_type >= 0):
            if (math.isnan(df.dist_bad[i]) or (df.dist_bad[i] > dist_from_cur)) and (dist_from_cur < df.dist_good[i]):
                df.dist_bad[i] = dist_from_cur
```

```python
num_angle = 360
small_decline = 0.0001
wpt = [55.61205, 39.97805]
cur = [55.59282, 39.96037]
fro = [55.59202, 39.95831]
cur_type = 0
free = 10 #km?
wpt[0] = float("{0:.6f}".format(wpt[0] - cur[0]))
wpt[1] = float("{0:.6f}".format(wpt[1] - cur[1]))
fro[0] = float("{0:.6f}".format(cur[0] - fro[0]))
fro[1] = float("{0:.6f}".format(cur[1] - fro[1]))
dist_to_wpt = rel_dist(wpt[0], wpt[1])
```

```python
%%time

df = pd.DataFrame(index = [i for i in range(num_angle)], columns = ['dist_bad', 'dist_good', 'angle_wpt', 'angle_from'])

file = 'N55bE039'
file_mp = file + '.mp'
m = open(file_mp).read()

N1 = m.count('[POLYLINE]')
t = 0
length_to_type = 16
length_to_coord = 7
for i in range(N1):
    t = m.find('[POLYLINE]', t + 1) + length_to_type
    type_line = int(m[t:m.find('\n', t)], 16)
    if not (type_line in polyline_good or type_line in polyline_bad):
        continue
    start = m.find('Data', t) + length_to_coord
    data = re.split('\),\(|,', m[start:(m.find('\n', start)-1)])
    j = 2
    length = len(data)
    while j < length:
        send_otr(float(data[j-2]), float(data[j-1]), float(data[j]), float(data[j+1]), type_line)
        j += 2
```
